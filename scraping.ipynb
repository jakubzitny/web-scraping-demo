# Quick examples

Let's try to parse something from webpage with a list of [158 MedTech startups](https://blog.benchsci.com/startups-using-artificial-intelligence-in-drug-discovery#aggregate_and_synthesize_information).
There is a list of companies that use AI to improve drug discovery R&D. The list includes info about each startup, including:
- Phase of the R&D
- Company name
- Company website
- Short description of company mission
- Year when the company was founded
- Place of headquarters

There is no Semantic Web info, however some of the data is easily parsable from the HTML DOM.

## Bash

````python
URL="https://blog.benchsci.com/startups-using-artificial-intelligence-in-drug-discovery#aggregate_and_synthesize_information"
!echo $URL
````

### Let's use `curl` to fetch the HTML of the webpage.

````python
!curl -s $URL
````

We can see there are some `h3` and  `h4` elements with phase, company names and urls.

### Let's try to filter only these, using `grep`:

````python
!curl -s $URL | grep h3
````

### Let's get filter the HTML tags using `sed`:

````python
!curl -s $URL | grep h3 | sed 's/^.*a>//g;s/<.*$//g' | tee categories.txt
````

### Something similar for the company names and URLs, inside `h4` tags:

````python
!curl -s $URL | grep h4
````

````python
!curl -s $URL | grep h4 | sed 's/^.*">//g;s/<.*$//g' | tee names.txt
````

### And extract the URLs from the `<a>` tags:

````python
!curl -s $URL | grep h4 | sed 's/^.*https/https/g;s/\".*$//g' | tee links.txt
````

We can use `wc` to check the number of lines in both `links.txt` and `names.txt` tags to see if the list is complete according to the webpage.

````python
!wc -l links.txt names.txt
````

### We can even use `paste` to merge the links and names together:

````python
!paste -d" " names.txt links.txt
````

## Node.js

The JS above was a jQuery-esque hacky dev-tooly solution. Let's introduce a better one in Node with `request`.

```js
const request = require('request')
const cheerio = require('cheerio')

const url = 'https://blog.benchsci.com/startups-using-artificial-intelligence-in-drug-discovery#aggregate_and_synthesize_information'

const handleResponse = (error, response, rawBody) => {
  if (error) {
    console.error('Failed to fetch the URL.')
    return
  }

  const $ = cheerio.load(rawBody)
  
  const h3s = $('h3').children().toArray()
    
  const categories = h3s.map((h3) => h3.next.data)

  console.log(categories)
  
  // and similarly for names and categories
}

request.get(url, handleResponse)
```


Alternative libraries for Node/JS are `https`, `axios`, `superagent` or `got`.

We'll be scraping posts from https://blog.scrapinghub.com.

````python
import scrapy

class BlogSpider(scrapy.Spider):
    name = 'blogspider'
    start_urls = ['https://blog.scrapinghub.com']

    def parse(self, response):
        for title in response.css('.post-header>h2'):
            yield {'title': title.css('a ::text').get()}

        for next_page in response.css('a.next-posts-link'):
            yield response.follow(next_page, self.parse)
````
